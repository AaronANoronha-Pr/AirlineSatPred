```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
```

```python
data.dropna(inplace=True)
```

```python
for satisfaction_level in data['satisfaction'].unique():
    print(f"\nSatisfaction level: {satisfaction_level}")
    subset = data[data['satisfaction'] == satisfaction_level]
    numeric_columns = subset.select_dtypes(include=['float64', 'int64']).columns
    numeric_mean_values = subset[numeric_columns].mean()
    print(numeric_mean_values)
```

```python
plt.figure(figsize=(8, 6))
satisfied_data_type_of_travel = data[(data['satisfaction'] == 'satisfied')].groupby('Type of Travel').size()
satisfied_data_type_of_travel.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Pie Chart for Satisfied Customers (Type of Travel)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
satisfied_data_class = data[(data['satisfaction'] == 'satisfied')].groupby('Class').size()
satisfied_data_class.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Pie Chart for Satisfied Customers (Class)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
neutral_or_dissatisfied_data_type_of_travel = data[(data['satisfaction'] == 'neutral or dissatisfied')].groupby('Type of Travel').size()
neutral_or_dissatisfied_data_type_of_travel.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Pie Chart for Neutral or Dissatisfied Customers (Type of Travel)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
neutral_or_dissatisfied_data_class = data[(data['satisfaction'] == 'neutral or dissatisfied')].groupby('Class').size()
neutral_or_dissatisfied_data_class.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Pie Chart for Neutral or Dissatisfied Customers (Class)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
